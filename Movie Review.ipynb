```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv("Datafiniti_Hotel_Reviews.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
x=df['reviews.text']
```

```python
y=df['reviews.rating']
y = np.where(df['reviews.rating']>3,1,0)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vec = CountVectorizer(min_df=5,ngram_range=(1,5)).fit(x_train)
```

```python
vec.get_feature_names()
```

```python
vecv = vec.transform(x_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
y_train = np.where(y_train>3,1,0)
```

```python
model.fit(vecv,y_train)
```

```python
fnames  = np.array(vec.get_feature_names())
```

```python
mc = model.coef_[0].argsort()
```

```python
fnames[mc[:10]]
```

```python
fnames[mc[-11:-1]]
```
